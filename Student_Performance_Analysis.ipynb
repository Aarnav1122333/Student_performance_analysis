**Objective:**  
To analyze student performance data and identify key factors influencing academic outcomes using exploratory data analysis.


```python
import pandas as pd
import matplotlib.pyplot as plt
```

importing of the dataset

```python
df = pd.read_csv('study_performance.csv')
```

```python
df
```

```python
df['parental_level_of_education'].value_counts()
```

**Finding mean of each students score**

```python
df['average_score']= df[['math_score','reading_score','writing_score']].mean(axis=1)
```

```python
df
```

**Loading data only related to female students**

```python
female_data=df[(df['gender']=='female')]
```

```python
female_data
```

**Sorting female students with respect to average score in decending order**

```python
female_data.sort_values('average_score',ascending=False)
```

**Loading only male data**

```python
male_data = df[(df['gender']=="male")]
```

```python
male_data
```

**Sorting male data with respect to average score in decending order**

```python
male_data.sort_values('average_score',ascending=False)
```

```python
male_data.sort_values('average_score',ascending=False).reset_index()
```

```python
female_data.sort_values('average_score',ascending=False).reset_index()
```

**Descritive statastics for male student data**

```python
male_data.describe()
```

**Descritive statastics for female student data**

```python
female_data.describe()
```

**This histogram compares the average scores of students across different subjects, highlighting subjects with higher or lower performance levels**

```python
plt.hist(df['average_score'], bins=10)
plt.xlabel('Score')
plt.ylabel('Number of Students')
plt.title('Distribution of Student Scores')
plt.show()
```
